```python
import tensorflow as tf
```

```python
scalar = tf.constant(7)
print(scalar, end="\n\n")

vector = tf.constant([5, 2, 6])
print(vector, end="\n\n")

matrix = tf.constant([[1, 2, 3] , [2, 5, 6]])
print(matrix, end="\n\n")

tensor = tf.constant([
    [[1, 2] , [3, 4]],
    [[5, 6],  [7, 8]]
])
print(tensor, end="\n\n")
```

```python
def tensor_info(name, t):
  print(f"{name}: \n Shape: {t.shape}\n Rank: {t.ndim}\n Size: {tf.size(t).numpy()}\n")
```

```python
tensor_info("Scalar", scalar)
tensor_info("Vector", vector)
tensor_info("Matrix", matrix)
tensor_info("Tensor", tensor)
```
