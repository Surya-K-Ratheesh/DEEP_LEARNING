```python
#One-hot encode the tensor you created in 9
```

```python
import tensorflow as tf
```

```python
tensor = tf.constant([3, 17, 5, 8, 2, 25, 14, 9, 6, 1])

depth = tf.reduce_max(tensor).numpy() + 1
```

```python
one_hot_encoded = tf.one_hot(tensor, depth=depth)
```

```python
print("Shape of one-hot encoded tensor:", one_hot_encoded.shape)
print("One-hot encoded tensor:\n", one_hot_encoded)
```
