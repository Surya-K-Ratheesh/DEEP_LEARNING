```python
#Multiply the two tensors (which you created for Question 3) using dot product.
```

```python
import tensorflow as tf
```

```python
tensor1 = tf.random.uniform(shape = [5, 300], minval = 0, maxval = 1)
tensor2 = tf.random.uniform(shape = [5, 300], minval = 0, maxval = 1)
```

```python
dot_product = tf.reduce_sum(tf.multiply(tensor1, tensor2), axis = 1)
```

```python
print(f"DOT PRODUCT SHAPE: {dot_product.shape}")
print(f"DOT PRODUCT RESULT: \n{dot_product}")
```
