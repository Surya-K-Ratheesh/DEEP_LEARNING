```python
#1. Create a vector, scalar, matrix and tensor with values of your choice using tf.constant()
```

```python
import tensorflow as tf
```

```python
#Scalar
scalar = tf.constant(7)
print(scalar)
```

```python
#Vector
vector = tf.constant([5, 2, 6])
print(vector)
```

```python
#Matrix
matrix = tf.constant([[1, 2, 3] , [2, 5, 6]])
print(matrix)
```

```python
#Tensor
tensor = tf.constant([
    [[1, 2] , [3, 4]],
    [[5, 6],  [7, 8]]
])
print(tensor)
```
