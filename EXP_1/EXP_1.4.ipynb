```python
#Multiply the two tensors (which you created for Question 3) using matrix multiplication.
```

```python
import tensorflow as tf
```

```python
tensor1 = tf.random.uniform(shape = [5, 300], minval = 0, maxval = 1)
tensor2 = tf.random.uniform(shape = [5, 300], minval = 0, maxval = 1)
```

```python
tensor2_T = tf.transpose(tensor2)
```

```python
result = tf.matmul(tensor1, tensor2_T)

print(f"RESULTANT SHAPE: {result.shape}")
print(f"RESULT: {result}")
```
