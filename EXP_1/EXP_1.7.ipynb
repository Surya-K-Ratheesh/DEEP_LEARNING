```python
#Find the min and max values of the tensor (which you created for Question 6)
```

```python
import tensorflow as tf
```

```python
tensor = tf.random.uniform(shape=[224, 224, 3], minval=0, maxval=1)
```

```python
min_val = tf.reduce_min(tensor).numpy()
max_val = tf.reduce_max(tensor).numpy()
```

```python
print(f"Minimum value in tensor: , {min_val}")
print(f"Maximum value in tensor: , {max_val}")
```
